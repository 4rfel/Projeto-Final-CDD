```python
import pandas as pd
data = pd.read_csv("olist_public_dataset_v2.csv")
data.set_index('order_id')
data.head()
```

```python
lista_classificacao= []
for i in data.review_score:
    
    if i == 5 or i == 4:
        classificacao = 1
        
    if i ==3:
        classificacao = 0
        
    if i == 2 or i == 1:
        classificacao = -1
        
    lista_classificacao.append(classificacao)
    
data["Classificacao"]=lista_classificacao
data['delivery_date'] = pd.to_datetime(data['order_delivered_customer_date'])
data['aproved_date'] = pd.to_datetime(data['order_aproved_at'])
data['delta time'] = data['delivery_date'] - data['aproved_date']
data['delta_days'] = (data['delta time'].dt.days)
```

```python
states_numbers = []
for state in data['costumer_state']:
    if state == 'SP':
        states_numbers.append(1)
    if state == 'RJ':
        states_numbers.append(2)
    if state == 'MG':
        states_numbers.append(3)
    if state == 'RS':
        states_numbers.append(4)
    if state == 'PR':
        states_numbers.append(5)
    if state == 'SC':
        states_numbers.append(6)
    if state == 'BA':
        states_numbers.append(7)
    if state == 'DF':
        states_numbers.append(8)
    if state == 'ES':
        states_numbers.append(9)
    if state == 'GO':
        states_numbers.append(10)
    if state == 'PE':
        states_numbers.append(11)
    if state == 'CE':
        states_numbers.append(12)
    if state == 'PA':
        states_numbers.append(13)
    if state == 'MT':
        states_numbers.append(14)
    if state == 'MA':
        states_numbers.append(15)
    if state == 'MS':
        states_numbers.append(16)
    if state == 'PB':
        states_numbers.append(17)
    if state == 'RN':
        states_numbers.append(18)
    if state == 'PI':
        states_numbers.append(19)
    if state == 'AL':
        states_numbers.append(20)
    if state == 'SE':
        states_numbers.append(21)
    if state == 'RO':
        states_numbers.append(22)
    if state == 'TO':
        states_numbers.append(23)
    if state == 'AM':
        states_numbers.append(24)
    if state == 'AC':
        states_numbers.append(25)
    if state == 'AP':
        states_numbers.append(26)
    if state == 'RR':
        states_numbers.append(27)
        
data["states_numbers"] = states_numbers
```

```python
data['customer_state'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

df = data[['order_products_value', 'order_freight_value', 'product_description_lenght', 'product_photos_qty', 'review_score', 'review_comment_message', 'delta_days', 'Classificacao', 'customer_state']]
df = df.dropna()

print(len(df))
dados_treinamento, dados_teste = train_test_split(df, test_size=0.4)
print(len(dados_treinamento))
print(len(dados_teste))
```

```python
y_train= dados_treinamento.Classificacao
x_train= dados_treinamento.review_comment_message
x_test= dados_teste.review_comment_message
y_test= dados_teste.Classificacao
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

# Cria modelo.
model = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("classifier", MultinomialNB())
])

# Treina modelo.
model.fit(x_train, y_train)
```

```python
# Testa modelo.
pos_prob = []
neg_prob = []
neu_prob = []
probs = model.predict_proba(x_train)
for e in probs:
    neg_prob.append(e[0])
    neu_prob.append(e[1])
    pos_prob.append(e[2])
dados_treinamento['pos_prob'] = pos_prob
dados_treinamento['neg_prob'] = neg_prob
dados_treinamento['neu_prob'] = neu_prob
dados_treinamento.head()


pos_prob = []
neg_prob = []
neu_prob = []
probs = model.predict_proba(x_test)
for e in probs:
    neg_prob.append(e[0])
    neu_prob.append(e[1])
    pos_prob.append(e[2])
dados_teste['pos_prob'] = pos_prob
dados_teste['neg_prob'] = neg_prob
dados_teste['neu_prob'] = neu_prob
```

```python
dados_treinamento.to_csv('train_data.csv')
dados_teste.to_csv('test_data.csv')
```
